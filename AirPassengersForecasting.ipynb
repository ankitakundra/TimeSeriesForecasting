## TIME SERIES FORECASTING

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.pyplot import rcParams
rcParams['figure.figsize'] = 10,6

from statsmodels.tsa.stattools import adfuller
#remove warnings
import warnings
def ignore_warn(*args, **kwargs):
    pass
warnings.warn = ignore_warn

```

```python
#we use date_parser to read the date column as datetime and not as String

dateparse = lambda x: pd.datetime.strptime(x, '%Y-%m')
data = pd.read_csv("AirPassengers.csv", parse_dates=['Month'], date_parser=dateparse)

# to set month as index
dataset = data.set_index(['Month'])
dataset.head(5)
```

```python
dataset.tail(5)
```

#### As we can see our data starts from the year 1949 to 1960

```python
# plotting the data

sns.tsplot(dataset['#Passengers'], time = dataset.index,color = "indianred")
```

#### we can see from the graph that the data is not stationary, since the mean that will be calculated at 1952 will be different from the mean at 1958. Therefore, we will be using methods to make the data stationary.

### Test to help to test the stationarity of the data
#### 1. Rolling Statistics
#### 2. Dickey-Fuller Test
     

```python
#Rolling Statistics
rolling_mean = dataset.rolling(window = 12).mean()
rolling_sd = dataset.rolling(window = 12).std()
print(rolling_mean)
```

#### Here the value of first 11th row is Nan, and the value from 12th row is the average of 1/12, 13th row is 2/13 and so on.

```python
orig= plt.plot(dataset,color = "blue",label = "original")
rmean = plt.plot(rolling_mean,color = "red",label = "rolling_mean")
rsd = plt.plot(rolling_sd,color = "black",label = "rolling_sd")
plt.legend(loc = "best")
```

#### As visible from the graph, the mean(red) and standard deviation (black) are not constant, it tends to be increasing which states that the data is not stationary. 

#### Now we will be testing the data using a 2nd method called ADFuller test. Here we have passed autog as AIC.

```python
dftest = adfuller(dataset['#Passengers'],autolag = 'AIC')

dfoutput = pd.Series(dftest[0:4], index = ['Test Statistics','p-value','#lags used','No of Observation used'])
for key,value in dftest[4].items():
     dfoutput['Critical Value (%s)'%key] = value

print(dfoutput)
```

#### The p-value is also high (it should be somewhat around 0.05) , critical value should also be more than Test Statistics, therefore we cannot reject null hypothesis and we can say that the data is non-stationary

#### Therefore we will take the log Scale of the dataset and check whether there is any change in the trend

```python
dataset_logScale = np.log(dataset)
plt.plot(dataset_logScale)
```

#### Again we will perform the Rolling Statistics test

```python
movingAverage = dataset_logScale.rolling(window = 12).mean()
movindSD = dataset_logScale.rolling(window = 12).std()

plt.plot(dataset_logScale)
plt.plot(movingAverage,color = "red")
```

#### As we can observe from the graph, the trend remains the same but it is better than the previous one. To make the data stationary, we will take the difference between the log value and the moving average.

```python
dataset_logScaleMinusMovingAverage = dataset_logScale - movingAverage
dataset_logScaleMinusMovingAverage.head(10)

#remove Nan Values
dataset_logScaleMinusMovingAverage.dropna(inplace = True)
dataset_logScaleMinusMovingAverage.head(10)
```

```python
#performing the rolling Statistics Test
dataset_logScaleMinusMovingAverage_mean = dataset_logScaleMinusMovingAverage.rolling(window =12).mean()
dataset_logScaleMinusMovingAverage_sd = dataset_logScaleMinusMovingAverage.rolling(window = 12).std()


plt.plot(dataset_logScaleMinusMovingAverage,color = "blue", label = "original log minus moving average")
plt.plot(dataset_logScaleMinusMovingAverage_mean,color = "red", label = "log minus moving average rolling mean")
plt.plot(dataset_logScaleMinusMovingAverage_sd,color = "black", label = "log minus moving average rolling sd")
plt.legend(loc = "best")
```

#### we can see from the graph that the mean and standard deviation has somewhat become constant, stating that the data is now stationary

```python
#ADFullter test
dftest = adfuller(dataset_logScaleMinusMovingAverage['#Passengers'],autolag = 'AIC')

dfoutput = pd.Series(dftest[0:4], index = ['Test Statistics','p-value','#lags used','No of Observation used'])
for key,value in dftest[4].items():
     dfoutput['Critical Value (%s)'%key] = value

print(dfoutput)
```

#### we see that the p-value is also relatively less and the Critical Value is also more than the Test Statistics value so we can say that the data is stationary

```python
#creating a function for rolling statistics and ADFullter test
def test_Stationarity(timeseries):
    
    
    movingAverageMean = timeseries.rolling(window = 12).mean()
    movingSD = timeseries.rolling(window = 12).std()
    
    orig= plt.plot(timeseries,color = "blue",label = "original")
    rmean = plt.plot(movingAverageMean,color = "red",label = "rolling_mean")
    rsd = plt.plot(movingSD,color = "black",label = "rolling_sd")
    plt.legend(loc = "best")
    
    dftest = adfuller(timeseries['#Passengers'],autolag = 'AIC')

    dfoutput = pd.Series(dftest[0:4], index = ['Test Statistics','p-value','#lags used','No of Observation used'])
    for key,value in dftest[4].items():
         dfoutput['Critical Value (%s)'%key] = value

    print(dfoutput)


    
```

```python
# now we have calculated the weighted average of the time series
exponentailDecayWeightedAverage = dataset_logScale.ewm(halflife = 12,min_periods = 0,adjust = True).mean()
plt.plot(dataset_logScale)
plt.plot(exponentailDecayWeightedAverage,color = "red")
```

```python
# substracting the log scale and the weighted Average
dataset_logScaleMinusMovingexpoDecayWeightedAverage = dataset_logScale - exponentailDecayWeightedAverage
test_Stationarity(dataset_logScaleMinusMovingexpoDecayWeightedAverage)
```

```python
#shifting the value in time series. taking a lag of 1
dataset_logdiffShifting = dataset_logScale - dataset_logScale.shift()
plt.plot(dataset_logdiffShifting)
```

```python
dataset_logdiffShifting.dropna(inplace = True)
#running the rolling statistics and ADFuller test
test_Stationarity(dataset_logdiffShifting)
```

### Components of Time series 

#### 1. Trend

#### 2. Cycle

#### 3. Seasonality

#### 4. Irregularities

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(dataset_logScale)

trend = decomposition.trend
seasonality = decomposition.seasonal
residual = decomposition.resid

plt.subplot(411)
plt.plot(dataset_logScale, label = "Original")
plt.legend(loc = "best")

plt.subplot(412)
plt.plot(trend,label = "Trend")
plt.legend(loc = "best")

plt.subplot(413)
plt.plot(seasonality,label = "Seasonality")
plt.legend(loc = "best")


plt.subplot(414)
plt.plot(residual,label = "Residuals")
plt.legend(loc = "best")

plt.tight_layout()



```

#### we can see that there is an upward trend which is linear in nature

#### Residual are nothing but irregularity in your data they do have any shape and size

```python
# testing whether the noise is stationary or not
decomposedLogData = residual
decomposedLogData.dropna(inplace = True)
test_Stationarity(decomposedLogData)
```

```python
#ACF and PACF to get the value of p and q value

from statsmodels.tsa.stattools import acf,pacf

#ols is ordinary least square
lag_acf = acf(dataset_logdiffShifting,nlags = 20)
lag_pacf = pacf(dataset_logdiffShifting, nlags = 20,method = 'ols')

#Plot ACF

plt.subplot(121)
plt.plot(lag_acf)
plt.axhline(y = 0,linestyle = '--',color = "gray")
plt.axhline(y = -1.96/np.sqrt(len(dataset_logdiffShifting)),linestyle = '--',color = "gray")
plt.axhline(y = 1.96/np.sqrt(len(dataset_logdiffShifting)),linestyle = '--',color = "gray")
plt.title('AutoCorrelation Function')


#plot PACF

plt.subplot(122)
plt.plot(lag_pacf)
plt.axhline(y = 0,linestyle = '--',color = "gray")
plt.axhline(y = -1.96/np.sqrt(len(dataset_logdiffShifting)),linestyle = '--',color = "gray")
plt.axhline(y = 1.96/np.sqrt(len(dataset_logdiffShifting)),linestyle = '--',color = "gray")
plt.title('Partial AutoCorrelation Function')

plt.tight_layout()
```

#### From the graph we need to at the value the graph drops to zero for the first time in both the ACF and PACF test. From the ACF, we get the value of p as 2 and from the PACF graph we can take the value of q as 2 

```python
from statsmodels.tsa.arima_model import  ARIMA
model  = ARIMA(dataset_logScale, order = (2,1,2))
results_AR = model.fit(disp = -1)
plt.plot(dataset_logdiffShifting)
plt.plot(results_AR.fittedvalues, color = "red")

#calculated RSS
plt.title("RSS: %.4f" % sum((results_AR.fittedvalues -dataset_logdiffShifting['#Passengers'])**2))
print('Plotting AR Model')
```

#### The value of RSS is good which is 1.0292

#### You can also play around the value of p and q

```python
#MA model
model  = ARIMA(dataset_logScale, order = (0,1,2))
results_AR = model.fit(disp = -1)
plt.plot(dataset_logdiffShifting)
plt.plot(results_AR.fittedvalues, color = "red")

plt.title("RSS: %.4f" % sum((results_AR.fittedvalues -dataset_logdiffShifting['#Passengers'])**2))
print('Plotting AR Model')
```

#### But we can see that value of RSS has increased which is not good. Similarily to get the AR model we need to take value of p = 2, d = 1 and q = 0

```python
# taking the old value of p, q and d
model  = ARIMA(dataset_logScale, order = (2,1,2))
results_ARIMA = model.fit(disp = -1)
plt.plot(dataset_logdiffShifting)
plt.plot(results_AR.fittedvalues, color = "red")

plt.title("RSS: %.4f" % sum((results_AR.fittedvalues -dataset_logdiffShifting['#Passengers'])**2))
print('Plotting AR Model')
```

```python
#Here we have converted the resultant ARIMA value into a series format and printed the head value
predictions_ARIMA_diff = pd.Series(results_ARIMA.fittedvalues,copy = True)
print(predictions_ARIMA_diff.head())
```

```python
#converting it to a cumulative sum
predictions_ARIMA_diff_cumsum = predictions_ARIMA_diff.cumsum()
print(predictions_ARIMA_diff_cumsum.head())
```

```python
#prediction done for the fitted value
predictions_ARIMA_log = pd.Series(dataset_logScale['#Passengers'].ix[0],index = dataset_logScale.index)
predictions_ARIMA_log = predictions_ARIMA_log.add(predictions_ARIMA_diff_cumsum,fill_value = 0)

predictions_ARIMA_log.head()
```

#### We do an exponentiation of the data so that it comes back to the original form

```python
predictions_ARIMA = np.exp(predictions_ARIMA_log)
plt.plot(dataset)
plt.plot(predictions_ARIMA)
```

#### Orange line indicates our fitted model, we can see that the magnitude is different but the shape is properly captured by the ARIMA model
